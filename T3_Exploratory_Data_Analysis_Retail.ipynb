```python
#Converting of object to float
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Count Plot (Category, Segment, Ship Mode, Region)

```python

fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(20,10));

sns.countplot(df['Category'],ax=axs[0][0],color='blue')
sns.countplot(df['Segment'],ax=axs[0][1],color='green')
sns.countplot(df['Ship Mode'],ax=axs[1][0],color='purple')
sns.countplot(df['Region'],ax=axs[1][1],color='orange')
axs[0][0].set_title('Category')
axs[0][1].set_title('Segment')
axs[1][0].set_title('Ship Mode')
axs[1][1].set_title('Region')
plt.tight_layout()
plt.show()
```

Showing of the Distribution of the Plot

```python

plt.figure(figsize=(12,8))
#Sales
plt.subplot(2,2,1)
sns.kdeplot(df['Sales'],color='red')
plt.title("Sales Distribution")
#Profit
plt.subplot(2, 2, 2)
sns.kdeplot( df['Profit'], color='green')
plt.title('Profit Distribution')
#Quantity
plt.subplot(2,2,3)
sns.kdeplot(df['Quantity'],color='blue')
plt.title("Quantity Distribution")
#Discounts
plt.subplot(2,2,4)
sns.kdeplot(df['Discount'],color='brown')
plt.title("Discount Distribution");
plt.tight_layout()
plt.show()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```
